```python
curr = input("Enter You Currency")
curr = curr.upper()
amount = input("Enter Amount")
amount = float(amount)

base_url = "https://api.exchangeratesapi.io/v1/latest?access_key=d7e10d7c99a750e0400b1c1fce15e9e8"+"&symbols="+curr
response = requests.get(base_url)

if(response.ok is False):
    print('\nError : {}'.format(response.status_code))
    print(response.json()['error'])
else:
    data = response.json()
    rate = data['rates'][curr]
    result = amount/rate

    print('{} {} as of today would be {} {}'.format(amount,curr,result, data['base']))
    
    
```

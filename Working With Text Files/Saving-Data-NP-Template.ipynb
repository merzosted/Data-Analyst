## Saving Files with NumPy

```python
import numpy as np
```

### np.save()

```python
lending_co = np.genfromtxt('Lending-company.csv', delimiter=',', dtype= np.str_)
lending_co
```

```python
np.save("Lending-Npy", lending_co)
```

```python
lending_data = np.load("Lending-Npy.npy")
lending_data
```

```python
np.array_equal(lending_co, lending_data)
```

### np.savez()

```python
lending_co = np.genfromtxt('Lending-Company-Saving.csv', delimiter=',', dtype= np.str_)
lending_data = np.load("Lending-Npy.npy")
```

```python
np.savez("Lending-Npy-Compressed", lending_co, lending_data)
```

```python
lending_data_savez = np.load("Lending-Npy-Compressed.npz")
print(lending_data_savez)
```

```python
lending_data_savez['arr_0']
```

```python
lending_data_savez['arr_1']
```

```python
np.savez("Lending-Comp-Npz", company=lending_co, saving=lending_data)
```

```python
npz_new = np.load("Lending-Comp-Npz.npz")
```

```python
npz_new.files
```
